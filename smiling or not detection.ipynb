```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import load_img,img_to_array
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

model=load_model('smiling or not.h5')
```

```python
image =load_img('not_smile_test_image.jpg',target_size=(124,124))
plt.imshow(image)
plt.show()
```

```python
image=img_to_array(image)
image=image.reshape(1,124,124,3)
```

```python
result= model.predict(image)
result
```

```python
result=np.argmax(result)
if result==0:
    print('Not smile')
else:
    print('smile')
```
